## OUTLIERS ESTIMATION FOR WICKETS

```python
percentile25 = df['Wkts'].quantile(0.25)
percentile75 = df['Wkts'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
iqr = percentile75 - percentile25
iqr
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
upper_limit
```

```python
lower_limit
```

## DETECTING OUTLIERS

```python
df[df['Wkts'] > upper_limit]
```

```python
df[df['Wkts'] < lower_limit]
```

## TRIMMING

```python
df3 = df[df['Wkts'] < upper_limit]
df3
```
